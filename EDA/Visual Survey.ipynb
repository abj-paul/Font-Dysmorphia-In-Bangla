```python
import os
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

def display_text_with_font(text, font_path):
    # Create a figure and axis
    fig, ax = plt.subplots()
    
    # Set the font properties using the specified TTF font file
    font_prop = FontProperties(fname=font_path)
    
    # Display the text on the axis with the specified font
    ax.text(0.5, 0.5, text, fontsize=12, fontproperties=font_prop, ha='center', va='center')
    
    # Remove axis ticks and labels for better visualization
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Show the plot
    plt.show()

# Specify the folder containing TTF font files
fonts_folder = '../fonts/'

# Specify the text to display
b_text_to_display = '''
এক অদ্বিতীয় সুপ্রসিদ্ধ বাঙালি কবি ও গল্পকার জয়ন্ত মন্ত্রীর অসীম আকাশে অবিচ্ছেদ্য বিশ্বাস।
The quick brown fox jumped over the lazy dog.
'''

text_to_display = '''
আমি বাংলায় খুব ভালোবাসি এবং এটি আমার মাতৃভাষা। আমার বাড়ি বহুত সুন্দর এবং আমি সেখানে খুব খুশি থাকি। সকাল বেলায় আমি পার্কে বই পড়তে যাই এবং সাথে কফি পান করি।
'''

# Iterate through all TTF font files in the fonts folder
for font_file in os.listdir(fonts_folder):
    if font_file.endswith(".ttf"):
        font_path = os.path.join(fonts_folder, font_file)
        
        # Call the function to display the text with the specified font
        display_text_with_font(text_to_display, font_path)
```
