```python
import os
import pandas as pd

# Set the path to the folder containing the CSV files
root_folder = "../data/"

# List to store dataframes from each CSV file
font_list = {}
dataframes = []

# Iterate over CSV files in the folder
for csv_file in os.listdir(root_folder):
    if csv_file.endswith(".csv"):
        csv_path = os.path.join(root_folder, csv_file)
        df = pd.read_csv(csv_path)
        font_list[csv_file.split(".csv")[0]] = df
        dataframes.append(df)

print(font_list)
```

```python
# Concatenate all dataframes into a single dataframe
full_df = pd.concat(dataframes, ignore_index=True)

# Display the combined dataframe
df = full_df
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_data(df):
    # Assuming 'full_df' is the combined dataframe from the previous code
    # You may need to adjust the column names based on your actual data
    font_name_col = 'Font'
    width_col = 'Width'
    height_col = 'Height'
    
    # Scatter plot for width vs height
    sns.scatterplot(x=width_col, y=height_col, hue=font_name_col, data=df, palette='viridis')
    plt.title('Width vs Height for Fonts')
    plt.xlabel('Width')
    plt.ylabel('Height')

    plt.show()
```
